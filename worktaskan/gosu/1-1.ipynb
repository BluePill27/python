```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
data = pd.read_json('table.jsonl', lines=True).dropna(subset=[['slot','x','y']])
```

```python
def Plot(slot, x, y):
    #extent=[65,185,65,185]
        
    fig, ax = plt.subplots()
    im = ax.hexbin(x, y,
                   gridsize=(25,10), bins='log', 
                   cmap=cm.jet, mincnt=1)
    fig.colorbar(im, ax=ax).set_label('log')
    
    #plt.axis([185,65,65,185])
    
    fig.savefig('plots/heatmap_' + str(slot)+  '.png', 
                dpi=fig.dpi, bbox_inches='tight', 
                pad_inches=0.5)
    
    plt.show()

for slot in data['slot'].unique():
    Plot(int(slot), data[data['slot'] == slot]['x'].as_matrix() ,data[data['slot'] == slot]['y'].as_matrix())
```
