```python
class Queue:
    def __init__(self): self.elements = {}
    def empty(self): return len(self.elements) == 0
    def put(self, item, priority): self.elements[item] = priority

    def get(self):
        best_item, best_priority = None, None
        for item, priority in self.elements.items():
            if best_priority is None or priority < best_priority:
                best_item, best_priority = item, priority

        del self.elements[best_item]
        return best_item, best_priority

class Graph:
    def __init__(self):
        self.edges = {}
        self.weights = {}
        self.costs = {}
        self.vertxes = []

    def add(self, a, b, w):
        self.edges.setdefault(a, []).append(b)
        self.weights[(a, b)] = w
        
    def neighbors(self, id): return self.edges.get(id, [])
    def cost(self, a, b): return self.weights.get((a, b), 1)
```

```python
def compute_path(graph, start, dest=None, tmp_graph=None):
    fron = Queue()
    fron.put(start, 0)
    
    came_from = {}
    cost_so_far = {}
    
    came_from[start] = None
    
    cost_so_far[start] = 0
    dest_cost = 0

    while not fron.empty():
        current, bp = fron.get()
        if current == dest:
            dest_cost = bp
        
        for next in graph.neighbors(current):
            new_cost = cost_so_far[current] + graph.cost(current, next)
            print ('New cost:', new_cost)
            if next not in cost_so_far or new_cost < cost_so_far[next]:
                cost_so_far[next] = new_cost
                fron.put(next, new_cost)
                
                if tmp_graph is not None:

                    if new_cost % 2 == 0: 
                        tmp_graph.weights[(current, next)] = 3
                    else:
                        tmp_graph.weights[(current, next)] = 1
                        
                came_from[next] = current
    print ('Des_cost', dest_cost)
    return dest_cost
```

```python
f     = open('input.txt',  'r')
f_out = open('output.txt', 'w')


v_numb = int(f.readline()) 
e_numb = int(f.readline())

s = int(f.readline())
t = int(f.readline())

graph = {}

for i in range(v_numb): 
    graph[i] = []

for line in f:
    # Split on any whitespace (including tab characters)
    row = line.split()
    # Convert strings to numeric values:
    i = int(row[0])
    connected_v  = int(row[1])
    graph[i].append(connected_v); 

#Processing the path
visited_nodes_flag = [False]*len(graph);


path = 1
unvisited_nodes= graph[s]
visited_nodes_flag[s]   = True 
discovered_nodes = [] 


while len(unvisited_nodes):


    del discovered_nodes[:]

    for node in unvisited_nodes:

        #check if the node was visited before
        if visited_nodes_flag[node]:
            continue

        #if not add it to visited   
        visited_nodes_flag[node] = True

        #check if the node is the target node
        if node == t :
            if path % 2 == 0:
                weigh = 3*path/2
            else:
                weigh = 3*(path-1)/2 + 1;

            f_out.write('%d' % weigh )

            f.close 
            f_out.close
            exit()
  
        #add its nieghbors
        discovered_nodes.extend(graph[node]) 

     
        #increse the path length
        path +=1
   
    del unvisited_nodes[:]

    #save discovered nodes
    unvisited_nodes.extend(discovered_nodes)
```
